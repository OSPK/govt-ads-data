```python
import pandas as pd

df = pd.read_csv('data.csv', index_col='Channel name')
# df.set_index('Channel name', inplace = True)
# print(list(df['Channel name']), list(df['2013-2014']))
# df['2013-2014']
# ax = df.plot.bar(x='2013-2014', y='Channel name', rot=0)
channels = list(df.index)

# for channel in channels:
#     df.loc[channel].plot(legend=True)
#plot usa column

```

```python
import pygal                                                       # First import pygal
from ipywidgets import HTML
from IPython.display import HTML
import base64
bar_chart = pygal.Bar(value_formatter=lambda x: 'PKR {:,.3f} mn'.format(x/1_000_000))
bar_chart.width = 1200
bar_chart.height = 600
bar_chart.x_labels = list(df.head())
bar_chart.show_y_guides=False
html_pygal = u"""
    <!DOCTYPE html>
    <html>
        <head>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
        </head>
        <body><figure><embed width='1200' height='600' src={pygal_render}></embed></figure></body>
    </html>
"""
for channel in channels[0:25]:
    bar_chart.add(channel, df.loc[channel])  # Add some values
bar_chart.render_to_file('bar-chart-by-paper.svg')
HTML(html_pygal.format(pygal_render=bar_chart.render_data_uri()))
```

```python
bar_chart = pygal.Line(interpolate='hermite', stroke_width='5px', value_formatter=lambda x: 'PKR {:,.3f} mn'.format(x/1_000_000))
bar_chart.x_labels = list(df.head())
bar_chart.width = 1200
bar_chart.height = 600
bar_chart.show_y_guides=False
bar_chart.legend_at_bottom=False
html_pygal = u"""
    <!DOCTYPE html>
    <html>
        <head>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
        </head>
        <body><figure><embed width='1200' height='500' src={pygal_render}></embed></figure></body>
    </html>
"""

bar_chart.add('Geo News', df.loc['Geo News'])  # Add some values
bar_chart.add('ARY News', df.loc['ARY News'], stroke_style={'width': 5})  # Add some values
bar_chart.add('Dunya News', df.loc['Dunya News'])  # Add some values
bar_chart.add('Express News', df.loc['Express News'])  # Add some values
bar_chart.add('Capital Tv', df.loc['Capital Tv'])  # Add some values
bar_chart.add('Dawn News', df.loc['Dawn News'])  # Add some values
bar_chart.add('Samma Tv', df.loc['Samma Tv'])  # Add some values

bar_chart.render_to_file('line-chart-by-paper.svg')
HTML(html_pygal.format(pygal_render=bar_chart.render_data_uri()))
```

```python
bar_chart = pygal.StackedLine(fill=True, value_formatter=lambda x: 'PKR {:,.3f} mn'.format(x/1_000_000))
bar_chart.x_labels = list(df.head())
bar_chart.width = 1200
bar_chart.height = 600
bar_chart.show_y_guides=False
bar_chart.legend_at_bottom=True
html_pygal = u"""
    <!DOCTYPE html>
    <html>
        <head>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
        </head>
        <body><figure><embed width='1200' height="500" src={pygal_render}></embed></figure></body>
    </html>
"""
for channel in channels[:25]:
    bar_chart.add(channel, df.loc[channel])  # Add some values

bar_chart.render_to_file('area-chart-by-paper.svg')
HTML(html_pygal.format(pygal_render=bar_chart.render_data_uri()))
```

```python
# sum(list(df['2013-2014']))
chart = pygal.Bar(value_formatter=lambda x: 'PKR {:,.3f} bn'.format(x/1_000_000_000))
chart.width = 1200
chart.height = 500
html_pygal = u"""
    <!DOCTYPE html>
    <html>
        <head>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
        </head>
        <body><figure><embed width='1200' height="500" src={pygal_render}></embed></figure></body>
    </html>
"""
years = list(df.head())
sums = []
chart.x_labels = years
chart.show_legend = False
for idx,year in enumerate(years):
    if idx > 4:
        color = '#2d84ef'
    else:
        color = '#d2ba4a'
    sums.append({'value': sum(list(df[year])), 'color':color })
chart.add('', sums)
chart.render_to_file('chartvs.svg')
HTML(html_pygal.format(pygal_render=chart.render_data_uri()))
```

```python
# sum(list(df['2013-2014']))
pie_chart = pygal.Treemap(value_formatter=lambda x: 'PKR {:,.3f} mn'.format(x/1_000_000))
pie_chart.width = 1200
pie_chart.height = 500
pie_chart.legend_at_bottom = True
html_pygal = u"""
    <!DOCTYPE html>
    <html>
        <head>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
            <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/pygal-tooltips.js"></script>
        </head>
        <body><figure><embed width='1200' height="500" src={pygal_render}></embed></figure></body>
    </html>
"""
for channel in channels:
    pie_chart.add(channel, sum(list(df.loc[channel])))
pie_chart.render_to_file('pie-chart-by-paper.svg')
HTML(html_pygal.format(pygal_render=pie_chart.render_data_uri()))
```
